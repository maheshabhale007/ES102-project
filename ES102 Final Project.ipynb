**Data Visualization**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#dataset1 : Men test match batting stats.
data1 = pd.read_csv(r'C:\Users\Mahesh Abhale\OneDrive\Desktop\ES102 Final Project\Men Test Team Batting Stats.csv')
#print(data1)

#dataset2 : Men test match bowling stats.
data2 = pd.read_csv(r'C:\Users\Mahesh Abhale\OneDrive\Desktop\ES102 Final Project\Men Test Team Bowling Stats.csv')
#print(data2)

#dataset3 : Men ODI match batting stats.
data3 = pd.read_csv(r'C:\Users\Mahesh Abhale\OneDrive\Desktop\ES102 Final Project\Men ODI Team Batting Stats.csv')
#print(data3)

#dataset4 : Men ODI match bowling stats.
data4 = pd.read_csv(r'C:\Users\Mahesh Abhale\OneDrive\Desktop\ES102 Final Project\Men ODI Team Bowling Stats.csv')
#print(data4)

```

**TEST MATCH DATA VISUALIZATION**


```python
print('TEST MATCH DATA VISUALISATION')
print()
print('BATTING STATS')
#print('MATCHES PLAYED BY EACH COUNTRY V/S MATCHES WON BY EACH COUNTRY')

print()
print('By the below graph shows the Matches played by each Country and the no. of matches won by each Country. The Countries are      represented by coloured dots to get clear idea.')
print()

plot1 = sns.relplot(data=data1,x='Team Matches Played',y='Matches Won',hue='Country')
plt.title('MATCHES PLAYED BY EACH COUNTRY V/S MATCHES WON BY EACH COUNTRY')
plt.show()

```

```python
#print('MATCHES WON BY EACH COUNTRY')
print()
print('The below scatterplot represents the "Matches won" by each Country along the X-axis. It also gives the information about the   "Win/Loss Ratio" of each Country which is represented by coloured dots and tells about the "Average Runs Per Six Balls Batting"  of each Country represented by different shapes.')

print()

plot2 = sns.relplot(x='Matches Won',y='Country',hue='Win/Loss Ratio',style='Avg Runs Per Six Balls Batting',palette=('ch:r=0.9,l=0.9'),data=data1)
plt.title('MATCHES WON BY EACH COUNTRY')
plt.show()
```

```python
#print('HIGHEST TEAM SCORE BATTING V/S AVERAGE RUNS PER WICKET BATTING')
print()

print('The line plot shown depicts the "Highest Score Of the Team" in relation with the "Avg Runs" they scored per wicket Batting.')

print()


plot3 = sns.relplot(x='Highest Team Score Batting',y='Avg Runs Per Wicket Batting',kind=('line'),data=data1)
plt.title('HIGHEST TEAM SCORE BATTING V/S AVERAGE RUNS PER WICKET BATTING')
plt.show()
```

```python
#print('MATCHES DRAWN V/S COUNTRY')
print()
print('The below sctterplot gives idea about the Matches Drawn along the X-axis and the different colour of the dots gives the sense   of Matches Tied.')

print()
plot4 = sns.catplot(x='Matches Drawn',y='Country',hue='Matches Tied',data=data1,s=7)
plt.title('MATCHES DRAWN V/S COUNTRY')
plt.show()
```

```python
#print('TEAM MATCHES PLAYED V/S MATCHES WON')
print()
print('In the following scatterplot five parameters are shown. "Team Matches Played" along X-axis, "Matches Won" along Y-axis,         "Matches Lost" by the coloured dots, "Win/Loss Ratio" by the size of dots and "Matches Drawn" by the shapes.')


plot5 = sns.relplot(data=data1,x='Team Matches Played',y='Matches Won',hue='Matches Lost',style='Matches Drawn',size='Win/Loss Ratio',palette=('ch:r=1,l=0.9'))
plt.title('TEAM MATCHES PLAYED V/S MATCHES WON')
plt.show()
```

```python
#print('Avg Runs Per Wicket Batting v/s Avg Runs Per Six Balls Batting')
print()
print('The plot below gives the information about "Avg Runs Per Wicket Batting" along X-axis and "Avg Runs Per Six Balls Batting"        along Y-axis')
print()

plot6 = sns.catplot(data=data1,x='Avg Runs Per Wicket Batting',y='Avg Runs Per Six Balls Batting',kind='point',color='r')
plot6.fig.autofmt_xdate()
plt.title('Avg Runs Per Wicket Batting v/s Avg Runs Per Six Balls Batting')
plt.show()
```

```python
#print('Highest Team Score Batting v/s Number Of Team Innings Batting')
print()
print('The scatterplot below gives an idea about the number of innings each country team batted and the highest score achieved by each country team with lowest completed score while batting shown by coloured dots.')
print()
plot7 = sns.relplot(data=data1,x='Number Of Team Innings Batting',y='Highest Team Score Batting',hue='Lowest Completed Score Batting',s=100,palette=('ch:r=1,l=0.7'))
plt.title('Highest Team Score Batting v/s Number Of Team Innings Batting')
plt.show()
```

```python
print('TEST MATCH DATA VISUALISATION')
print()
print('BOWLING STATS')
print()
print('The barplot shows the "Avg Runs Per Wicket Bowling" along the Y-axis by each Country team on the X-axis.')
print()
plot8 = sns.catplot(x='Country',y='Avg Runs Per Wicket Bowling',kind='bar',data=data2)
#plt.rc('figure',figsize=(40,10))
plt.title('Avg Runs Per Wicket Bowling v/s Country')
plot8.fig.autofmt_xdate()
plt.show()
```

```python
print('PAIRWISE VISUALISATION OF DATA')
print()
print('The below pairplot gives a clear view of all each data compared to the other.')
print()
plot9 = sns.pairplot(data=data2,height=4)
plot9.fig.autofmt_xdate()
plot9.map_offdiag(sns.lineplot)
plt.show()
```

**ODI MATCH DATA VISUALIZATION**

```python
print('ODI MATCH DATA VISUALISATION')
print()
print('BATTING STATS')
print()

print('The line graph below shows the "Team Matches Played" along Y-axis and "Country" along X-axis by the green line and            "Matches Won" along Y-axis and "country" along X-axis by red line.')
print()
#plt.figure(figsize=(40,20))
plot10 = plt.plot(data3['Country'],data3['Team Matches Played'],linewidth=5,color='g')
plot11 = plt.plot(data3['Country'],data3['Matches Won'],linewidth=5,color='r')
plt.xlabel('Country')
plt.ylabel('Team Matches Played  /   Matches Won')
plt.title('(Team Matches Played  /  Matches Won)  v/s Country')
plt.legend(['Team Matches Played','Matches Won'])
plt.show()
```

```python
print()
print('The bar plot below shows the Win/Loss Ratio of each country team')
print()

plot12 = plt.bar(data3['Country'],data3['Win/Loss Ratio'],color='m',width=0.6)
plt.xlabel('Country')
plt.ylabel('Win/Loss Ratio')
plt.title('Win/Loss Ratio  v/s  Country')
plt.show()
```

```python
print()
print('This plot gives an idea about "Highest Team Score Batting  v/s  Number Of Team Innings Batting"')
print()

plot13 = sns.catplot(data=data3,x='Number Of Team Innings Batting',y='Highest Team Score Batting',kind='bar')
plot13.fig.autofmt_xdate()
plt.title('Highest Team Score Batting  v/s  Number Of Team Innings Batting')
plt.show()
```

```python


plot14 = sns.relplot(x='Avg Runs Per Six Balls Batting',y='Avg Runs Per Wicket Batting',hue='Country',data=data3,s=50)
plt.title('Avg Runs Per Wicket Batting  v/s  Avg Runs Per Six Balls Batting')
plt.show()
```
